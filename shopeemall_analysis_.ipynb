```python
import pandas as pd 
from snownlp import SnowNLP


data = pd.read_csv('shopeemall_comment_analysis.csv')
data
```

```python
semantic = []
for row in data.itertuples():##將商品評價表個中每一列讀取
    s = SnowNLP(row.comment)
    ##呼叫sentiments屬性評價情緒分數
    
    if s.sentiments > 0.5:
        semantic.append('正面')
    else:
        semantic.append('負面')

    # print(s.sentiments)
data['semantic'] = semantic

data
```

```python
##將同一個商品資料分別加總正負評價
result = data.groupby('title')['semantic'].value_counts().to_frame()
result
```
